```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

# Dados Trabalho

```python
dados_trabalho_2005_df = pd.read_csv('dados_trabalho_2005.csv', sep=';')
dados_trabalho_2005_df.head(100)
```

```python
dados_trabalho_2005_df.fillna(0, inplace=True)
dados_trabalho_2005_df
```

```python
dados_trabalho_2015_df = pd.read_csv('dados_trabalho_2015.csv', sep=';')
dados_trabalho_2015_df.head(11)
```

```python
dados_trabalho_2015_df.fillna(0, inplace=True)
dados_trabalho_2015_df
```

## Análise da distribuição das classes

### 2005 vs 2015

```python
cols1 = ['Homens_10_anos', 'Mulheres_10_anos']
cols2 = ['Homens_10_anos', 'Mulheres_10_anos']
desc_cols = ['entre Homens (10 anos)', 'entre Mulheres (10 anos)']
years = [2005, 2015]
dfs = [dados_trabalho_2005_df, dados_trabalho_2015_df]

num_cols = 2
fig, ax = plt.subplots(nrows=1, ncols=num_cols, figsize=(20, num_cols*5))

for index, col in enumerate(dfs):
   
    height1 = dfs[index][cols1[0]]
    height2 = dfs[index][cols1[1]]

    bars = dados_trabalho_2005_df['Classes']
    y_pos = np.arange(len(bars))
    plt.figure(figsize=(20, 7))
    
    p1 = ax[index].barh(y_pos, height1, alpha = 0.8, color = 'b')
    p2 = ax[index].barh(y_pos, height2, alpha = 0.3, color = 'r')
    
    ax[index].title.set_text(f"Distribuição dos salários mínimos em ({years[index]})")
    ax[index].set_yticks(y_pos)
    ax[index].set_yticklabels(bars)
    ax[index].set_xlabel('População impactada (1000)')
    ax[index].legend((p1[0], p2[0]), desc_cols)
```

#### Conclusão: houve uma significativa redução no número de pessoas sem rendimento, reduzindo a distância entre este número e a quantidade de pessoas que recebem entre 1 a 2 salários mínimos.
