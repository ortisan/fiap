```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

# Dados de Educação

```python
dados_educacao_2005_df = pd.read_csv('dados_educacao_2005.csv', sep=';')
dados_educacao_2005_df.head(100)
```

```python
dados_educacao_2005_df = dados_educacao_2005_df[:-1] #drop lastline 
dados_educacao_2005_df
```

```python
dados_educacao_2015_df = pd.read_csv('dados_educacao_2015.csv', sep=';')
dados_educacao_2015_df.head(100)
```

## Análise da distribuição das classes

### Comparação entre os totais entre 2005 e 2015

```python
desc_cols = ['2005', '2015']
height1 = dados_educacao_2005_df['Total']
height2 = dados_educacao_2015_df['Total']

bars = dados_educacao_2005_df['Classes']
y_pos = np.arange(len(bars))
plt.figure(figsize=(20, 10))
height = 0.4
plt.yticks(y_pos - height /2, bars)
p1 = plt.barh(y_pos, height1, height=height)
p2 = plt.barh(y_pos - height, height2, height=height)
plt.legend((p1[0], p2[0]), desc_cols)

plt.xlabel('Pessoas de 5 anos ou mais de idade (1000 pessoas)')
plt.title('Educação - Pessoas de 5 anos ou mais de idade (1000 pessoas)')
plt.show()
```

## Comparativo entre alfabetizados

```python
df_alfabetizados_2005_df = pd.read_csv('dados_educacao_alfabetizadas_2005.csv', sep=';')
df_alfabetizados_2005_df.head(100)
```

```python
df_alfabetizados_2005_df = df_alfabetizados_2005_df[:-1] #drop lastline 
df_alfabetizados_2005_df
```

```python
df_alfabetizados_2015_df = pd.read_csv('dados_educacao_alfabetizadas_2015.csv', sep=';')
df_alfabetizados_2015_df.head(100)
```

```python
desc_cols = ['2005', '2015']

df_alfabetizados_2005_df['Total_NORM'] = df_alfabetizados_2005_df['Total'] / df_alfabetizados_2005_df['Total'].sum()
df_alfabetizados_2015_df['Total_NORM'] = df_alfabetizados_2015_df['Total'] / df_alfabetizados_2015_df['Total'].sum()

height1 = df_alfabetizados_2005_df['Total_NORM']
height2 = df_alfabetizados_2015_df['Total_NORM']

bars = dados_educacao_2005_df['Classes']
y_pos = np.arange(len(bars))
plt.figure(figsize=(20, 10))
height = 0.4
plt.yticks(y_pos - height /2, bars)
p1 = plt.barh(y_pos, height1, height=height)
p2 = plt.barh(y_pos - height, height2, height=height)
plt.legend((p1[0], p2[0]), desc_cols)

plt.xlabel('Taxa de pessoas de 5 anos ou mais de idade alfabetizadas')
plt.title('Educação - Comparativo entre a taxa de pessoas de 5 anos ou mais de idade entre 2005 e 2015')
plt.show()
```

#### Conclusão: Entre as idades de 7 a 29 anos taxa de pessoas alfabetizadas diminuiu, e entre 30 a 60 anos a taxa aumentou.
