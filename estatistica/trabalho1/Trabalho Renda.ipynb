```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(20, 10))
```

# Dados Gerais

```python
dados_gerais_2005_df = pd.read_csv('dados_gerais_2005.csv', sep=';')
dados_gerais_2005_df.head(11)
```

```python
dados_gerais_2015_df = pd.read_csv('dados_gerais_2015.csv', sep=';')
dados_gerais_2015_df.head(11)
```

## Análise da distribuição das classes

```python
height = dados_gerais_2005_df['Total']
bars = dados_gerais_2005_df['Classes']
y_pos = np.arange(len(bars))
plt.figure(figsize=(20, 10))
plt.barh(y_pos, height)
plt.yticks(y_pos, bars)
plt.xlabel('População residente (1000)')
plt.title('Distribuição dos grupos de idades dos brasileiros(2005)')
plt.show()
```

#### Conclusão: Existe uma concentração maior da população entre 5 a 24 anos. 

## Distribição entre a zona urbana e rural

```python
cols = ['Total_URB', 'Total_RUR']
desc_cols = ['Urbana', 'Rural']

num_lines = len(cols)
fig, ax = plt.subplots(nrows=1, ncols=num_lines, figsize=(20, num_lines*5))

for index, col in enumerate(cols):
    height = dados_gerais_2005_df[col]
    bars = dados_gerais_2005_df['Classes']
    y_pos = np.arange(len(bars))
    
    ax[index].barh(y_pos, height)
    ax[index].title.set_text(f"Distribuição dos grupos de idades dos brasileiros da zona {desc_cols[index].lower()} (2005)")
    ax[index].set_yticks(y_pos)
    ax[index].set_yticklabels(bars)
    ax[index].set_xlabel('População residente (1000)')

plt.show()
```

#### Conclusão: guardadas as proporções, as distribuições são iguais

```python
height1 = dados_gerais_2005_df[cols[0]]
height2 = dados_gerais_2005_df[cols[1]]

bars = dados_gerais_2005_df['Classes']
y_pos = np.arange(len(bars))
plt.figure(figsize=(20, 10))

plt.yticks(y_pos, bars)
p1 = plt.barh(y_pos, height1)
p2 = plt.barh(y_pos, height2)
plt.legend((p1[0], p2[0]), desc_cols)

plt.xlabel('População residente (1000)')
plt.title('Distribuição dos grupos de idades dos brasileiros entre zona urbana e rural(2005)')
plt.show()
```

#### Conclusão: Grande concentração na zona urbana

## Distribição entre a zona urbana de homens e mulheres

```python
cols = ['Homens_URB', 'Mulheres_URB']
desc_cols = ['Homens', 'Mulheres']

height1 = dados_gerais_2005_df[cols[0]]
height2 = dados_gerais_2005_df[cols[1]]

bars = dados_gerais_2005_df['Classes']
y_pos = np.arange(len(bars))
plt.figure(figsize=(20, 10))

plt.yticks(y_pos, bars)
p1 = plt.barh(y_pos, height1, alpha = 0.8, color = 'b')
p2 = plt.barh(y_pos, height2, alpha = 0.3, color = 'r')
plt.legend((p1[0], p2[0]), desc_cols)

plt.xlabel('População residente (1000)')
plt.title('Distribuição dos grupos de idades entre homens e mulheres de zona urbana (2005)')
plt.show()
```

#### Conclusão: Na zona urbana, a pupulação de idade entre 0 e 14 anos são maioria homens. O restante - 15 a 70 anos - são mulheres
